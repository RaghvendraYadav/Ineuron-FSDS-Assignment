### Average weekly response time for each agent

```python
def Average_Response_Time(time_split):
    hh,mm,ss=time_split.split(':')
    return int(hh)*3600+int(mm)*60+int(ss)

                              
```

```python
df1['Average_Response_Time_Seconds']=df1['Average Response Time'].apply(Average_Response_Time)
```

```python
df1.groupby(['Agent Name','Data_week'])['Average_Response_Time_Seconds'].mean()
```
